<h1>Number operator in the energy basis</h1>

```python
import matplotlib.pyplot as plt
import numpy as np
from qutip import *

from two_coupled_qubits import ChargeQubit
```

```python
q = ChargeQubit(e_c=0.23084, e_j=16.34, levels=5, n_g=0)
_, eigenvectors = q.h.eigenstates()

elements = []
for i in range(q.dim):
    row = []
    for j in range(q.dim):
        row.append(np.abs((eigenvectors[i].dag() * q.number[1] * eigenvectors[j])[0][0][0]))
    elements.append(row)
```

```python
fig, ax = plt.subplots()

c = plt.pcolor(range(q.dim), range(q.dim), elements, cmap="inferno")
colorbar = fig.colorbar(c)

plt.xlabel("Hamiltonian eigenstates")
plt.ylabel("Hamiltonian eigenstates")
colorbar.set_label('N matrix elements')
plt.show()
```
